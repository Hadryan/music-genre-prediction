```python
# BUNDLE ARCHIVE
artists_data = pd.read_csv("../final-data/bundle_archive/artists-data.csv")
lyrics_data = pd.read_csv("../final-data/bundle_archive/lyrics-data.csv")
artists_data.head()
```

```python
artists_data['Artist'] = artists_data['Artist'].apply(lambda row: row.lower())
artists_data['Genre'] = artists_data['Genre'].apply(lambda row: row.lower())
artists_data['Genres'] = artists_data['Genres'].apply(lambda row: str(row).lower())
# artists_data.sort_values(by="Genre",ascending=True)
artists_data.sort_values(by="Artist",ascending=True,inplace=True)

artists_data.head()
```

```python
artists_data.drop_duplicates(subset=['Artist'],inplace=True)
artists_data.sort_values(by="Link",ascending=True).head()
```

```python
lyrics_data.head()
```

```python
lyrics_data['SName'] = lyrics_data['SName'].apply(lambda row: row.lower())
lyrics_data['Lyric'] = lyrics_data['Lyric'].apply(lambda row: str(row).lower())
lyrics_data['Idiom'] = lyrics_data['Idiom'].apply(lambda row: str(row).lower())
lyrics_data = lyrics_data[lyrics_data.Idiom == "english"]
lyrics_data.head()
```

```python
# add artists to lyrics df
lyrics_artist_data = pd.merge(lyrics_data, artists_data, how="left",left_on="ALink",right_on="Link")

# for some reason is creating duplicate rows with different Genre ?????
lyrics_artist_data.head(20)
```

```python
# lyrics_artist_data.drop("Link",1,inplace=True)
lyrics_artist_data.rename(columns={"Songs":"ASongs","Popularity":"APopularity","Genre":"AGenre","Genres":"AGenres"}, inplace=True)

for col in lyrics_artist_data.columns:
    lyrics_artist_data[col] = lyrics_artist_data[col].apply(lambda cell: str(cell))

lyrics_artist_data.head()
```

```python
len(lyrics_artist_data)
for artist in np.sort(lyrics_artist_data.Artist.unique()):
    print(artist)
```

```python
lyrics_artist_data.to_csv("bundle_lyrics_dataframe.csv",index=False,header=True)
```
